# The second part (Hyperparameter tuning)

```python
from sklearn.model_selection import GridSearchCV
```

```python
# following the grid parameters from the Qsar-co-x toolkit.
param_grid = {
    'bootstrap': [True, False],
    'max_depth': [10, 30, 50, 70, 90, 100, 200],
    'max_features': ['sqrt', 'log2'],
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10],
    'n_estimators': [50, 100, 200, 500],
    'criterion': ["gini", "entropy"]}
```

```python
# CV comes from the QSAR-co-x toolkit default parameter.
clf_tun = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 5, n_jobs = -1)
```

```python
clf_tun.fit(xtr,ytr)
```

```python
rf_best=clf_tun.best_estimator_
tree.plot_tree(rf_best.estimators_[0], feature_names=xtr.columns)
```

```python
print(str(clf_tun.best_estimator_))
```
